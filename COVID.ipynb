```python
import pandas as pd

# Fetch new data for each run.
vaccination_positive = pd.read_csv("https://onemocneni-aktualne.mzcr.cz/api/v2/covid-19/ockovani-pozitivni.csv")
vaccination_deaths = pd.read_csv("https://onemocneni-aktualne.mzcr.cz/api/v2/covid-19/ockovani-umrti.csv")
vaccination_hospitalization = pd.read_csv("https://onemocneni-aktualne.mzcr.cz/api/v2/covid-19/ockovani-hospitalizace.csv")
```

```python
# Turn 'datum' column into datetime
vaccination_positive['datum'] = pd.to_datetime(vaccination_positive['datum'], yearfirst=True)
vaccination_deaths['datum'] = pd.to_datetime(vaccination_deaths['datum'], yearfirst=True)
vaccination_hospitalization['datum'] = pd.to_datetime(vaccination_hospitalization['datum'], yearfirst=True)
```

```python
print(
    vaccination_positive['datum'][0],
    vaccination_deaths['datum'][0],
    vaccination_hospitalization['datum'][0]
)
```

```python
# Only taking in account last 'wave', which we will unofficially start from 01.09.2021
vaccination_positive = vaccination_positive[vaccination_positive['datum'] >= '2021-09-01'].head(vaccination_positive.shape[0] - 1)
vaccination_deaths = vaccination_deaths[vaccination_deaths['datum'] >= '2021-09-01'].head(vaccination_deaths.shape[0] - 1)
vaccination_hospitalization = vaccination_hospitalization[vaccination_hospitalization['datum'] >= '2021-09-01'].head(vaccination_hospitalization.shape[0] - 1)
```

```python
import matplotlib.pyplot as plt
# a little test

plt.figure(figsize=(14, 7))
plt.plot(vaccination_positive['datum'], vaccination_positive['pozitivni_dokoncene_ockovani_relativni_pocet'])

plt.title("COVID Pozitivni od zari.")
plt.xlabel('Datum')
plt.ylabel('Pozitivni celkem')

plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20, 10))

plt.plot(vaccination_positive['datum'], vaccination_positive['pozitivni_dokoncene_ockovani_relativni_pocet'], label="Pozitivni ockovani")
plt.plot(vaccination_hospitalization['datum'], vaccination_hospitalization['hospitalizovani_dokoncene_ockovani_relativni_pocet'], label="Hospitalizovani ockovani")
plt.plot(vaccination_deaths['datum'], vaccination_deaths['zemreli_dokoncene_ockovani_relativni_pocet'], label="Umrti ockovani")

plt.title("Relativni pocet ockovanych mezi nakazenymi, hospitalizacemi a umrtimi.")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 14.5))

plt.subplot(3, 1, 1)
plt.plot(vaccination_positive['datum'], vaccination_positive['pozitivni_bez_ockovani'], label='Bez ockovani')
plt.plot(vaccination_positive['datum'], vaccination_positive['pozitivni_nedokoncene_ockovani'], label='Nedokoncene ockovani')
plt.plot(vaccination_positive['datum'], vaccination_positive['pozitivni_dokoncene_ockovani'], label='Dokoncene ockovani')
plt.plot(vaccination_positive['datum'], vaccination_positive['pozitivni_posilujici_davka'], label='Treti davka')
plt.title("Pozitivni dle ockovani.")
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(vaccination_hospitalization['datum'], vaccination_hospitalization['hospitalizovani_bez_ockovani'], label='Bez ockovani')
plt.plot(vaccination_hospitalization['datum'], vaccination_hospitalization['hospitalizovani_nedokoncene_ockovani'], label='Nedokoncene ockovani')
plt.plot(vaccination_hospitalization['datum'], vaccination_hospitalization['hospitalizovani_dokoncene_ockovani'], label='Dokoncene ockovani')
plt.plot(vaccination_hospitalization['datum'], vaccination_hospitalization['hospitalizovani_posilujici_davka'], label='Treti davka')
plt.title("Hospitalizace dle ockovani.")
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(vaccination_deaths['datum'], vaccination_deaths['zemreli_bez_ockovani'], label='Bez ockovani')
plt.plot(vaccination_deaths['datum'], vaccination_deaths['zemreli_nedokoncene_ockovani'], label='Nedokoncene ockovani')
plt.plot(vaccination_deaths['datum'], vaccination_deaths['zemreli_dokoncene_ockovani'], label='Dokoncene ockovani')
plt.plot(vaccination_deaths['datum'], vaccination_deaths['zemreli_posilujici_davka'], label='Treti davka')
plt.title("Zemreli dle ockovani dle ockovani.")
plt.legend()

plt.show()
```
